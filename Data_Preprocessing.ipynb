```python
# Converting DType of Order_Date from object to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'] , format='mixed')
```

```python
# Checking what irrelavant values present in Time_Order_picked feature
invalid_mask_time_ordered = pd.to_datetime(df['Time_Orderd'], errors='coerce').isna()

wrong_values_to = df.loc[invalid_mask_time_ordered, 'Time_Orderd']
wrong_values_to.value_counts()
```

```python
# Function for time correction as there present 24:05 which is not a valid one so converting into 00:05
def correct_time(t):
    if pd.isna(t):
        return t
    if t.startswith("24:"):
        return "00:"+t.split(":")[1]
    return t

df['Time_Order_picked'] = df['Time_Order_picked'].apply(correct_time)
```

```python
# Checking what irrelavant values present in Time_Order_picked feature
invalid_mask_time_order_picked = pd.to_datetime(df['Time_Order_picked'], errors='coerce').isna()

wrong_values_top = df.loc[invalid_mask_time_order_picked, 'Time_Order_picked']
wrong_values_top.value_counts()
```

```python
# Converting Time_Ordered and Time_Order_picked feature to datetime and also removing irrelavant values like (0.875 , 0.75 , 1 , etc) using `errors = 'coerce'`

df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'],
            format="%H:%M",
            errors='coerce').dt.time

df['Time_Order_picked'] = pd.to_datetime(df['Time_Order_picked'],
            format="%H:%M",
            errors='coerce').dt.time
```

```python
# Creating order_datetime from Order_Date + Time_Ordered
order_datetime = pd.to_datetime(
    df['Order_Date'].astype('str') + " " + df['Time_Orderd'].astype('str'),
    errors='coerce'
)

df.insert(10 , "Order_Datetime" , order_datetime)

# Creating order_datetime from Order_Date + Time_Order_picked
Pickup_Datetime = pd.to_datetime(
    df['Order_Date'].astype('str') + " " + df['Time_Order_picked'].astype('str'),
    errors='coerce'
)

df.insert(11 , "Pickup_Datetime" , Pickup_Datetime) 
```

```python
# Some of the pickup done overnight (00:05 or 00:10)
mask = df['Pickup_Datetime'] < df['Order_Datetime']

df.loc[mask , 'Pickup_Datetime'] = df.loc[mask , 'Pickup_Datetime'] + pd.Timedelta(days = 1)
```

```python
# Droping Order_Date , Time_Ordered , Time_Order_Picked as we have designed new freatures from them
df.drop(columns=['Order_Date' , "Time_Orderd" , "Time_Order_picked"] , inplace=True)
```

```python
# Prep Time 
prep_time_min = ((df['Pickup_Datetime'] - df['Order_Datetime']).dt.total_seconds()/60)

df.insert(12 , "Prep_Time(min)" , prep_time_min)
```

```python
# Droping these rows because imputation is risky
df.drop(index=df[df['Order_Datetime'].isna() & df['Pickup_Datetime'].isna()].index  , inplace=True)
```

```python
# Statistical Information of Prep Time Feature
df['Prep_Time(min)'].describe()
```

```python
# Median Prep Time
median_prep_time = df['Prep_Time(min)'].median()

# imputing the values of Order_Datetime with Pickup_Time Order_Datetime - median 
mask1 = df['Order_Datetime'].isna() 
df.loc[mask1 , "Order_Datetime"] = df.loc[mask1 , "Pickup_Datetime"] - pd.Timedelta(minutes=median_prep_time)

# imputing the values of Pickup_time with Order_Datetime + median 
mask2 = df['Pickup_Datetime'].isna()
df.loc[mask2 , 'Pickup_Datetime'] = df.loc[mask2 , "Order_Datetime"] + pd.Timedelta(minutes=median_prep_time)
```

```python
# updating the Prep Time 
df['Prep_Time(min)'] = (df['Pickup_Datetime'] - df['Order_Datetime']).dt.total_seconds()/60
```

```python
# These Features are very important for both analysis and prediction and all of them contains null value and when imputing them with some mode (random way) will decrease the efficiency of the model and imputation will become unreliable 


df.drop(index=df[df['Delivery_person_Ratings'].isna() & df['Weather_conditions'].isna() & df['Road_traffic_density'].isna() & df['multiple_deliveries'].isna()].index , inplace=True)


df.drop(index=df[df['Delivery_person_Ratings'].isna() & df['Weather_conditions'].isna() & df['Road_traffic_density'].isna()].index , inplace=True)
```
